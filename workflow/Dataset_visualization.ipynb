```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import os
```

```python
df = pd.read_csv(
    '../dataset/styles.csv',
    on_bad_lines = 'skip'  
                  )

df = df.dropna()
df = pd.DataFrame(df)

image_path = r'../dataset/images'
```

```python
df.head(4)
```

Dataset visualization

```python
def show_k_random(k, name):
    choices = np.random.choice(df.index, k, replace=False)
    choices = df.loc[choices, ['id', name]]
    fig, axs = plt.subplots(1, k, figsize = (15, 10))
    for this_name, id, ax in zip(choices[name], choices['id'], axs.ravel()):
        img_path = f"{image_path}/{id}.jpg"
        img = plt.imread(img_path)
        ax.imshow(img, cmap='gray')
        ax.set_xticks([])
        ax.set_yticks([])
        ax.set_title(f'{this_name}')
    plt.tight_layout()
    plt.show()
```

```python
labels = df.columns[1:-1]
for col in labels:
    print(col)
    show_k_random(10, col)
```
